```python
dataFeatures = list(data.columns[:-1])
print("Features Columns: \n{}".format(dataFeatures))
targetFeature = data.columns[-1]
print("\nTarget Column: \n{}".format(targetFeature))
```

```python
featureValues = data[dataFeatures]
targetValues = data[targetFeature].replace(['yes', 'no'], [1, 0])
print ("\nFeature values: \n{}".format(featureValues.head()))
print("\nTarget Feature values: \n{}".format(targetValues.head()))
```

```python
def Preprocessing(X):
    df = pd.DataFrame(index=X.index)
    for column, columnData in X.iteritems():
        if columnData.dtype == object:
            columnData = columnData.replace(['yes', 'no'], [1, 0])
        if columnData.dtype == object:
            columnData = pd.get_dummies(columnData, prefix=column)
        df = df.join(columnData)
    return df

featureValues = Preprocessing(featureValues)
print ("Processed Features ({}):-\n{}".format(len(featureValues.columns), list(featureValues.columns)))
```

```python
allValues = data.shape[0]
print(allValues)
```

```python
from sklearn.model_selection import train_test_split
trainingSet = int(allValues/100*75)
testSet = allValues - trainingSet
```

```python
X_train, X_test, y_train, y_test = train_test_split(featureValues,targetValues,test_size = 0.25,train_size=0.75, random_state = 42)
print("Traning Set: {} samples".format(X_train.shape[0]))
print("Test Set: {} samples".format(X_test.shape[0]))
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logistic = LogisticRegression(random_state = 0)
logistic.fit(X_train,y_train)
```

```python
logistic.predict(X_test)
```

```python
lrAccuracy = logistic.score(X_test,y_test)*100
print("Logistic Regression Accuracy {:.2f}%".format(lrAccuracy))
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
K = range(1,21)
K_score = []
print("k\tAccuracy(%)")
for k in K:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train,y_train)
    score = knn.score(X_test, y_test)
    K_score.append(score)
    print(str(k)+"\t{:.2f}".format(score*100))
```

```python
maxAccuracyKNN = max(K_score)*100
print("Maximum Accuracy from KNeighbors Classifier {:.2f}%".format(maxAccuracyKNN))
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
NB = GaussianNB()
NB.fit(X_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
y_predicted = NB.predict(X_test)
NBaccuracy = accuracy_score(y_test,y_predicted)*100
print("Naive Bayes Accuracy: {:.2f}%".format(NBaccuracy))
```
